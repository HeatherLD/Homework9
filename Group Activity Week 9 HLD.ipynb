## 1a) What's the difference between linear and rank correlation?

linear correlation (uses Pearson's corr coeff): linear correlation shows how closely a set of data aligns with a linear function. A relationship between variables that is measured to be close to a linear function has a higher absolute value correlation coefficient.

rank correlation (Spearman's rho, Kendall coeff): compares data sets when they are assigned to ranks (i.e. 1st place, 2nd place, etc.) If the ranks of two or more data sets are similar, then their correlation is strong. The particular values of the data don't matter as much as their assigned rank in the system.

## 1c) Calculate the correlation coefficient and describe what is happening.

```python
belize_df["availability_365"].corr(belize_df["calculated_host_listings_count"])
```

The correlation coefficient of this scatterplot is about 0.173, which is low in the range of 0 to 1 (where 1 is the strongest correlation). Thus, the host_listings_count and availability_365 data do not seem to be related to each other or affect each other -- they seem fairly independent.

## 2) Create a box and whisker plot for price and explain it in simple terms.

```python
fig, ax = plt.subplots()
ax.boxplot([belize_df["price"]], vert=False)
ax.set_title("Range of Price in Dollars")
ax.set_xlabel("Price in Dollars")
plt.show()
```

The price range as depicted shows that the mean and IQR are fairly meaningless for this data, since there are so many outliers. The price of a stay ranges widely, due in part to the fact that some prices are calculated per night, and some prices are the result of a multiple-night stay entered as a lump sum.
